```python
df['HaveWorkedWith'] = df['HaveWorkedWith'].astype(str)
df['HaveWorkedWith']=[x.split(';') for x in df['HaveWorkedWith']]
df
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
mlb = MultiLabelBinarizer(sparse_output=True)
df=df.join(
            pd.DataFrame.sparse.from_spmatrix(
                mlb.fit_transform(df.pop('HaveWorkedWith')),
                index=df.index,
                columns=mlb.classes_))
df
```

```python
x=df.drop(["Employed"],axis=1)
y=df["Employed"].values
```

```python
plt.pie(df['EdLevel'].value_counts(),labels=["Undergraduate","Master","Other","NoHigherEd","PhD"])
plt.legend()
print(df['EdLevel'].value_counts())
```

```python
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder
o=make_column_transformer((OneHotEncoder(handle_unknown='ignore'),['EdLevel']),remainder='passthrough')
x=o.fit_transform(x)
x
```

```python
from sklearn.model_selection import train_test_split
xtr,xte,ytrain,ytest=train_test_split(x,y,test_size=0.30,random_state=42)
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc.fit(xtr)
xtrain=sc.transform(xtr)
xtest=sc.transform(xte)
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
k=KNeighborsClassifier(n_neighbors=5)
g=GaussianNB()
s=SVC()
dt=DecisionTreeClassifier(criterion='entropy')
rf=RandomForestClassifier(n_estimators=100)
l=[k,g,s,dt,rf]
for i in l:
  print("Model Name:",i)
  i.fit(xtrain,ytrain)
  ypred=i.predict(xtest)
  print("Accuracy Score=",accuracy_score(ytest,ypred))
  print("Confusion matrix=\n",confusion_matrix(ytest,ypred))
  print("Classification Record:\n",classification_report(ytest,ypred))
  print("*+"*15)
```
