```python
Q1=df.quantile(0.25)
Q3=df.quantile(0.75)
IQR=Q3-Q1
threshold=1.5
df=df[~((df<(Q1-threshold*IQR)) | (df>(Q3+threshold*IQR))).any(axis=1)]
df
```

```python
x_discrete=df[['city','statezip','country']]
x_continuous=df[['bedrooms','sqft_living','sqft_lot','floors','condition','sqft_above','sqft_basement','yr_built','yr_renovated']]
y_copy=df['price']
```

```python
from sklearn.feature_selection import f_classif
score=f_classif(x_continuous,y_copy)
score
```

```python
f_value=pd.Series(score[0],index=x_continuous.columns)
f_value.sort_values(ascending=False)
```

```python
p_value=pd.Series(score[1],index=x_continuous.columns)
p_value.sort_values(ascending=False)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
x_test
```

```python
y_train
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor()
param={'n_estimators':[100,110,120],'max_depth':[10,12,16],'max_features':[4,8,10]}
gd=GridSearchCV(rf,param,cv=10)
gd.fit(x_train,y_train)
```

```python
print(gd.best_params_)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error,mean_squared_error,r2_score
print('MAE is',mean_absolute_error(y_test,y_pred))
print('MAE is',mean_absolute_percentage_error(y_test,y_pred))
msq=mean_squared_error(y_test,y_pred)
print('MSE is',msq)
sq=np.sqrt(msq)
print('RMSE is',sq)
print('R2 score is',r2_score(y_test,y_pred))
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error,mean_squared_error,r2_score
dtr=DecisionTreeRegressor(max_depth=7,min_samples_split=4,min_samples_leaf=1)
rfr=RandomForestRegressor(n_estimators=120,max_depth=16,max_features=10)
model=[dtr,rfr]
```

```python
for i in model:
  print(i)
  i.fit(x_train,y_train)
  y_pred=i.predict(x_test)
  print('MAE is',mean_absolute_error(y_test,y_pred))
  print('MAPE is',mean_absolute_percentage_error(y_test,y_pred))
  msq=mean_squared_error(y_test,y_pred)
  print("MSE is",msq)
  sq=np.sqrt(msq)
  print('RMSE is',sq)
  print('R2 score is',r2_score(y_test,y_pred))
  print('*'*100)
```
